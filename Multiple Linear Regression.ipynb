Multiple Linear Regression

Importing The Librarires

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

Importing the Dataset

```python
df=pd.read_csv("50_Startups.csv")
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

Encoding Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
x
#! makes 3 new columns to encode the data like cities in the x
#! no need to do feature extraction in multiple linear regression
```

Test train split of DataSet

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Train the Multiple Linear Regression Model on Training Set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

Predicting the model accuracy

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
#* any value with only 2 decimal places
#? concatinating the 2 columns 
# takes tupple as a input 
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
#! reshape = row column
```

Checking the accuracy of the regression

```python
print('Train Score: ', regressor.score(x_train, y_train))
print('Test Score: ', regressor.score(x_test, y_test))
```

More Information about the Model

```python
print(regressor.intercept_)
print(regressor.coef_)
```

Finding The Scores of the Models

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

Plotting The Graphs

```python
plt.plot(y_test, color='red')
plt.plot(y_pred,color='blue')
plt.show()
```
